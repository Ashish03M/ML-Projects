## <a name='3'> Approach for the analysis of data, for finding out answers</a>

- **Goal :** Before finding out the approach , let's understand the goal of our analysis first.
    - Our goal here is to find out whether we can market Bluebull health drink as 'Height Enhancer' or not and some other related question, using the 2 sets of children under observation.
    - Generally , if we are going to label a product as 'Height Enhancer' , so the most basic condition it has to satisfy that it should increase the height , and here that product is "Bluebull Health Drink".
- **Approach :** Now as we have clear goal in mind , we understand that , we have to use the Height parameter for our analysis.
    - Now since we have 2 different datasets of 2 sets of Children under observation :
        - Control : Not consuming Bluebull Health Drink
        - Test : Consuming Bluebull HEalth Drink
    - We can compare the change in the height of these 2 groups ,or more specifically the mean change in height,and find out if there is any change in growth of both groups.
    - And to do this ,we'll use **The independent two-sample t-test/unpaired t-test** used in *Hypothesis Testing*.

## <a name='4'> Hypothesis Testing ???</a>

- A hypothesis test is a method of statistical inference used to decide
    whether the data at hand sufficiently support a particular hypothesis.
- Here we use 2 hypothesis to form the basis of our test:
    - Null Hypothesis(H0) : Nothing new is happening ,or no effect of the new experiment/drug/training etc.
    - Alternate Hypothesis(H1/Ha) : The alternative hypothesis, is a statement that contradicts the null hypothesis and claims
there is a significant effect of new experiment/medication/training etc.
- We do analysis and gather inferences to **reject null hypothesis**. 
- **Independent two-sample t-test/Unpaired t-test :** 
     - This type of hypothesis test is used to compare the means of two independent samples. 
     - The null hypothesis states that there is no significant difference between the means of the two samples, while the alternative hypothesis states that there is a significant difference.

## <a name='6'> Statistical Analysis and Hypothesis Testing</a>

- Since we have 2 different samples for comparison , we are using **Independent t-test/Unpaired t-test**.
- Now to use Independent t-test there are some **pre-requistes :**
    - **Both samples should be approximately normally distributed**. If the sample sizes are large (typically n > 30), the Central Limit Theorem often ensures that the sample means are approximately normally distributed, even if the underlying population distributions are not normal. In case if data is not distributed normally we'll use **mannwhitneyu test**.
    - The variances of the two groups being compared should be approximately equal. This assumption is called **homogeneity of variance**.If this assumption is not met, you can use *Welch's t-test*, which does not require equal variances.
    - **Independence of observations**: The two samples must be independent, meaning there is no relationship between the observations in one group and the observations in the othergroup. The subjects in the two groups should be selected randomly and independently.

### Homogeneity of Variance 

```python
sns.boxplot(x="Base Height", y='Group',data=data, palette=["skyblue", "orange"])
plt.show()
```

```python
sns.boxplot(x="Month 12", y='Group',data=data, palette=["skyblue", "orange"])
plt.show()
```

```python
statistic, p_value = levene(control['Base Height'],test['Base Height'])
print("Levene's Test Statistic:", statistic)
print("P-value:", p_value)

# Interpret the results
alpha = 0.05
if p_value < alpha:
    print("Variances are not equal.")
else:
    print("Variances are equal.") 
```

**Observation :**
- Box-Plot clearly shows the spread of data ,which is equal ,proves that the Variance of Height is same in both group
- Also , We can use **Levene's Test** , which is a statistical test used to assess the equality of variances for two or more groups. 
- Now since the data is *not normally distributed* and *Variance is equal* for Both group for Height , we'll apply *Mann-Whitney U Test* for Hypothesis Testing for safe side and accuracy. 
- Although it doesn't mean if data is not *normally distirbuted* we can't use *Independent T Test* all ,it also gives significant result ,but to be more accurate we avoid it ,and use *Mann-Whitney U Test*.
- We'll do both and check in terms of Accuracy .

```python
# Mann-Whitney U Test


stat, p_value = mannwhitneyu( control['Month 12'],test['Month 12'], alternative='less')
print("Mann-Whitney U statistic:", stat)
print("Mann-Whitney U test P-value:", round(p_value,4))

# Interpret the results
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: mean(height growth of Test) > mean(height growth of Control) ")
else:
    print("Fail to reject the null hypothesis: mean(height growth of Test) = mean(height growth of Control) ")
```

```python
# Indpendent T Test

stat, p_value = ttest_ind( control['Month 12'],test['Month 12'], alternative='less',equal_var=True)
print("ttest_ind statistic:", stat)
print("ttest_ind P-value:",round( p_value,4))

# Interpret the results
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis:mean(height growth of Test) > mean(height growth of Control) ")
else:
    print("Fail to reject the null hypothesis: mean(height growth of Test) = mean(height growth of Control)")
```

**Observation :**
- Comparing the two p-values 
    - Mann-Whitney U test P-value: 0.0091
    - ttest_ind P-value: 0.015
- We can confidently reject the null hypothesis ,but the *Mann Whitney p-value* is more significant and provides stronger evidence against the null hypothesis.
- In other words, if you are using a significance level (α) of 0.05, the result with p≈0.009 would lead to rejecting the null hypothesis, while the result with p≈0.015 might not be significant enough to reject the null hypothesis at the 0.05 level.
- Therefore it is preferable to use *Mann-Whitney U Test*.

```python
#As Bluebull is established as height enhancer , we'll use only use "Test" data to test the significance across states.
# Now Since we are using single sample for testing , we'll use 'Related T Test/Paired T Test'.


import warnings
warnings.filterwarnings('ignore')

significant_states = []
insignificant_states=[]
p=[]


for state in test['State'].unique():
    month_1_height = test[(test['State'] == state)]['Base Height']
    month_12_height = test[(test['State'] == state)]['Month 12']

    t_stat, p_value = ttest_rel(month_1_height, month_12_height)
    alpha = 0.05

    if p_value < alpha:
        p.append(p_value)
        significant_states.append(state)
    else :
        p.append(p_value)
        insignificant_states.append(state)

print("States where BlueBull can be marketed based on the experiment:", significant_states, '\n')
print("States where BlueBull need to work ,and can't be marketed for now:", insignificant_states)
```

```python

plt.figure(figsize=(17, 6))
sns.lineplot(x=test['State'].unique(),y=p,marker='o',markersize=10)
plt.xticks(rotation=45, ha='right', fontsize=10)  # Adjust rotation angle, alignment, and font size
plt.xlabel('State',fontsize=18)  
plt.ylabel('Count', fontsize=14) 
plt.title('p value across State', fontsize=20) 
plt.tight_layout() 
plt.show()

```

**Observation :**
- In Majority of the states , it is working pretty good as 'Height Enhancer ' and we can market it throught out India.
